```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import logistic_regression as lr
sns.set_style('darkgrid')
```

## First Dataset

```python
seed = 1280946712
# seed = np.random.randint(0, 1<<32)
print(seed)
np.random.seed(seed)
w = np.random.normal(size=2)
b = np.random.normal()

X = np.random.uniform(0, 1, size=(100, 2))
y = (lr.sigmoid(20*(X @ w + b)) < np.random.uniform(0, 1, X.shape[0])).astype(np.float32)

data_1 = pd.DataFrame()
data_1['x0'] = X[:,0]
data_1['x1'] = X[:,1]
data_1['y'] = y

data_1.to_csv('data_1.csv', index=False)

plt.figure(figsize=(5, 5))
sns.scatterplot(x='x0', y='x1', hue='y', data=data_1);
```

## Second Dataset

```python
seed = np.random.randint(0, 1<<32)
seed = 3782462388
print(seed)
np.random.seed(seed)

data2 = []

for split in ['train', 'test']:
    
    X = np.random.normal(0, 1, size=(500, 2))
    y_raw = lr.sigmoid(10*(np.linalg.norm(X, axis=-1) - np.sqrt(4/np.pi)))
    y = (y_raw < np.random.uniform(size=X.shape[0])).astype(np.float32)
    df = pd.DataFrame()
    df['x0'] = X[:, 0]
    df['x1'] = X[:, 1]
    df['y'] = y
    df['split'] = split
    data2.append(df)
    print(split, y.mean())
    
data2 = pd.concat(data2)
# data2.to_csv('data_2.csv', index=False)


data2_train = data2.query('split == "train"')
data2_test = data2.query('split == "test"')
fig, (ax1, ax2) = plt.subplots(figsize=(10, 5), ncols=2)
ax1.set_title('Train')
sns.scatterplot(x='x0', y='x1', hue='y', data=data2_train, ax=ax1)
ax2.set_title('Test')
sns.scatterplot(x='x0', y='x1', hue='y', data=data2_test, ax=ax2);
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set random seed for reproducibility
seed = np.random.randint(0, 1 << 32)
np.random.seed(seed)
print(seed)

# Generate random data points
num_samples = 1000

data2 = []

for split in ['train', 'test']:
    X = np.random.rand(500, 2)
    y = np.sin(X[:, 0] * np.pi) * np.sin(X[:, 1] * np.pi) + 0.1 * np.random.randn(500)
    y = (y > 0).astype(np.float32)
    
    df = pd.DataFrame()
    df['x0'] = X[:, 0]
    df['x1'] = X[:, 1]
    df['y'] = y
    df['split'] = split
    data2.append(df)
    print(split, y.mean())

data2 = pd.concat(data2)
# data2.to_csv('data_2.csv', index=False)

data2_train = data2.query('split == "train"')
data2_test = data2.query('split == "test"')
fig, (ax1, ax2) = plt.subplots(figsize=(10, 5), ncols=2)
ax1.set_title('Train')
sns.scatterplot(x='x0', y='x1', hue='y', data=data2_train, ax=ax1)
ax2.set_title('Test')
sns.scatterplot(x='x0', y='x1', hue='y', data=data2_test, ax=ax2)
plt.show()
```
